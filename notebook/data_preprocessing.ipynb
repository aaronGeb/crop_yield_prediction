```python
# select the unwanted columns
col = [
    "Unnamed: 0",
    "Field_ID",
    "Location",
    "Soil_fertility",
    "Pollution_level",
    "Plot_size",
    "Annual_yield",
]
```

```python
def drop_columns(data, columns_to_drop):
    """
    Removes specified columns from a DataFrame.

    Parameters:
    -----------
    data : pandas.DataFrame
        The input DataFrame from which columns need to be dropped.

    columns_to_drop : list
        A list of column names to be removed from the DataFrame.

    Returns:
    --------
    pandas.DataFrame
        A new DataFrame with the specified columns removed.

    Raises:
    -------
    KeyError:
        If any of the columns in `columns_to_drop` are not present in the DataFrame.
    """
    try:
        return data.drop(columns=columns_to_drop, axis=1)
    except KeyError as e:
        print(f"Error: {e}. Please check that all columns exist in the DataFrame.")
        raise
```

```python
for c in col:
    data = drop_columns(data, c)
```
